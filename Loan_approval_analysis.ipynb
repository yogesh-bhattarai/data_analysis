```python
#adding cibil rating column with values(poor,average,good,excellent)
def cibil_rating(value):
    if 300<=value<=549:
        return 'Poor'
    elif 550<=value<=649:
        return 'Average'
    elif 650<=value<=749:
        return 'Good'
    elif 750<=value<=900:
        return 'Excellent'
```

```python
df['cibil_rating']= df['cibil_score'].apply(cibil_rating)
```

```python
#grouping customers according to their income
def income_levels(value):
    if 100000<=value<=1000000:
        return "Low"
    elif 1000001<=value<=4000000:
        return 'Low-Middle'
    elif 4000001<=value<=7500000:
        return 'Upper-Middle'
    elif 7500001<=value<=10000000:
        return 'High'
```

```python
df['income_levels']= df['income_annum'].apply(income_levels)
```

```python
df.shape
```

```python
df['income_levels'].value_counts().plot(kind='bar')
```

```python
#grouping loan amounts as bronze,silver,gold and platinum
def loan_rating(value):
    if 100000<=value<=10000000:
        return 'Bronze'
    elif 10000001<=value<=20000000:
        return 'Silver'
    elif 20000001<=value<=30000000:
        return 'Gold'
    elif 30000001<=value<=40000000:
        return 'Platinum'
df['loan_rating']=df['loan_amount'].apply(loan_rating)
```

```python
#grouping  loan term as short,intermediate,long-term
def loan_term_label(value):
    if 0<=value<=7:
        return 'Short_Term'
    elif 8<=value<=14:
        return 'Intermediate'
    elif 15<=value<=20:
        return 'Long_Term'
df['loan_term_label']= df['loan_term'].apply(loan_term_label)
```

```python
df.head()
```

```python
df['loan_term_label'].value_counts().plot(kind='bar')
```

```python
#adding a dependency level column
def dependency(value):
    if 0<=value<=1:
        return 'Low'
    elif 2<=value<=3:
        return 'Moderate'
    elif 3<=value<=4:
        return 'High'
df['Dependent_levels']=df['no_of_dependents'].apply(dependency)  
        
```
