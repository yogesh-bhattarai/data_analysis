```python
#display daily sales day of the week
day={
    0:'Mon',
    1:'Tue',
    2:'Wed',
    3:'Thu',
    4:'Fri',
    5:'Sat',
    6:'Sun'
}
```

```python
#chaging date to datetime datatype
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df['day_of_week']=df['Date'].dt.dayofweek.map(day)
```

```python
df.head()
```

```python
df['day_of_week'].value_counts().plot(kind='bar')
```

```python
#week vs quantity
df.groupby('day_of_week')['Quantity'].mean().plot(kind='bar')
```
