```python
# Classify Movies Based on Ratings [Excellent, Good, and Average]
df['Rating']
```

```python
def Rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return"Good"
    else:
        return 'Bad'
```

```python
df['Rating_by_language']=df['Rating'].apply(Rating)
```
