```python
cast_shows=df[df['cast']!='No cast'].set_index('title').cast.str.split(',',expand=True).stack().reset_index(level=1,drop=True)
plt.figure(figsize=(10,5))
plt.title('Top 10 actor based on the no of the title')
sns.countplot(y=cast_shows,order=cast_shows.value_counts().index[:10])
plt.plot()
```

```python
movies_df=df[df['type']=='Movie']
len(movies_df)
```

```python
show_df=df[df['type']=='TV Show']
```

```python
movies_df
```

```python
movies_df['duration']=movies_df['duration'].apply(lambda x:x.replace('min','')if 'min' in x else x)
movies_df.loc[:,['duration']]=movies_df.loc[:,['duration']].apply(lambda x:x.astype(int))
```

```python
movies_df.info()
```

```python
#shortest movie
movies_df[movies_df['duration']==np.min(movies_df['duration'])]
```

```python
#longest movie
movies_df[movies_df['duration']==np.max(movies_df['duration'])]
```

```python
show_df
```

```python
show_df['duration']=show_df['duration'].apply(lambda x:x.replace('Season','')if 'Season' in x else x)
```

```python
show_df
```

```python
show_df['duration']=show_df['duration'].apply(lambda x:x.replace('s','')if 's' in x else x)
show_df.loc[:,['duration']]=show_df.loc[:,['duration']].apply(lambda x:x.astype(int))
```

```python
#show the highest number of seasons
longest_shows=show_df.loc[(show_df.duration>=13)]
```
